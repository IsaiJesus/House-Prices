```python
def create_scrolleable_table(df, table_id, title):
    html = f'<h3>{title}</h3>'
    html += f'<div id="{table_id}" style="height: 220px; overflow: auto;">'
    html += df.to_html()
    html += '</div>'
    return html
```

# 2. Exploratory Data Analysis (EDA) & Data Cleaning

### Missing Data

```python
count_missing_values = df_train.isnull().sum()

missing_values = count_missing_values.reset_index()
missing_values.columns = ['Column', 'Quantity']
missing_values = missing_values[missing_values['Quantity'] > 0]
missing_values = missing_values.sort_values(by='Quantity', ascending=False)
```

```python
html_missing_values = create_scrolleable_table(missing_values, 'missing_values', 'Missing values')

display(HTML(html_missing_values))
```

```python
# Delete values with more than 50% of the total data

total_rows = df_train.shape[0]

columns_to_delete = count_missing_values[count_missing_values > (total_rows * 0.5)].index

df_train = df_train.drop(columns=columns_to_delete)
```

```python
df_train.shape
```

```python
# Fill null number values with its mean 

columns_to_fill = count_missing_values[count_missing_values <= (total_rows * 0.5)].index
columns_to_fill = df_train[columns_to_fill].select_dtypes(include=['number']).columns

for col in columns_to_fill:
    mean = df_train[col].mean()
    df_train[col] = df_train[col].fillna(mean)
```

```python
# Fill null object values with its top

count_missing_values = df_train.isnull().sum()

columns_to_fill = count_missing_values.index
columns_to_fill = df_train[columns_to_fill].select_dtypes(include=[object]).columns

for col in columns_to_fill:
    top_value = df_train[col].value_counts().idxmax()
    df_train[col] = df_train[col].fillna(top_value)
```

```python
null_cols = df_train.isnull().any().sum()
print(f"Columnas con nulos: {null_cols}")
```

### Data transformation

At this point, I was thinking about transforming all the categorical values, but I didn't know what the best way to do it would be, so I asked ChatGPT how I could proceed and this is its answer.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

def custom_features(df):
    df = df.copy()
    
    # Binary
    df['CentralAir'] = df['CentralAir'].map({
        'N': 0,
        'Y': 1
    }).astype(int)
    
    df['PavedDrive'] = df['PavedDrive'].map({
        'N': 0,
        'Y': 1,
        'P': 2
    }).astype(int)
    
    # Nominal
    nominal_features = [
        'MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Condition1', 'Condition2', 
        'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'Foundation', 
        'BsmtFinType1', 'BsmtFinType2', 'Heating', 'Electrical', 'Functional', 'GarageType', 'GarageFinish', 
        'SaleType', 'SaleCondition'
    ]

    encoder = LabelEncoder()

    for n_feature in nominal_features:
        df[n_feature] = encoder.fit_transform(df[n_feature])
        
    freq_neighborhood = df['Neighborhood'].value_counts()    # Frecuency Encoding

    df['Neighborhood'] = df['Neighborhood'].map(freq_neighborhood)
    
    # Ordinal
    categorical_order = [['NoSeWa', 'AllPub']]    # Smallest to largest

    encoder = OrdinalEncoder(categories=categorical_order)

    df['Utilities'] = encoder.fit_transform(df[['Utilities']])
    
    ordinal_features = [
        'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC', 'KitchenQual', 
        'FireplaceQu', 'GarageQual', 'GarageCond'
    ]

    categorical_order = [['Po', 'Fa', 'TA', 'Gd',   'Ex']]

    encoder = OrdinalEncoder(categories=categorical_order)

    for o_feature in ordinal_features:
        df[o_feature] = encoder.fit_transform(df[[o_feature]])
    
    categorical_order = [['No', 'Av', 'Mn', 'Gd']]

    encoder = OrdinalEncoder(categories=categorical_order)

    df['BsmtExposure'] = encoder.fit_transform(df[['BsmtExposure']])
```

# 4. Feature Engineering 

### 1. Create new features

```python
# Total Square Footage

df_train['TotalSF'] = df_train['TotalBsmtSF'] + df_train['1stFlrSF'] + df_train['2ndFlrSF']
```

```python
# Quality-to-Price Ratio

df_train['QualityPriceRatio'] = df_train['OverallQual'] / df_train['SalePrice']
```

```python
# House Age

df_train['HouseAge'] = df_train['YrSold'] - df_train['YearBuilt']
```

```python
# Time Since Last Remodel

df_train['SinceRemodel'] = df_train['YrSold'] - df_train['YearRemodAdd']
```

### 2. Generate Binary Features

```python
# Does it have a basement, garage, or pool?

df_train['HasBsmt'] = (df_train['TotalBsmtSF'] > 0).astype(int)
df_train['HasGarage'] = (df_train['GarageArea'] > 0).astype(int)
df_train['HasPool'] = (df_train['PoolArea'] > 0).astype(int)
```

### Correlations

```python
corr = df_train.corr()['SalePrice'].drop('SalePrice')

# High positive correlations
positive_corr = corr[corr > 0.6].sort_values(ascending=False)
print("Variables with high positive correlation:")
print(positive_corr, '\n')

# High negative correlations
negative_corr = corr[corr < -0.6].sort_values(ascending=False)
print("Variables with high negative correlation:")
print(negative_corr)
```

### 7. Validate Your New Features

```python
new_features = df_train.head()
html_new_features = create_scrolleable_table(new_features, 'transformed_features', 'Transformed features')

display(HTML(html_new_features))
```

```python
features = positive_corr.index.tolist() + negative_corr.index.tolist()
y = 'SalePrice'

n_rows = (len(features) + 1) // 2
fig, axes = plt.subplots(n_rows, 2, figsize=(10, n_rows * 3), sharey=True)
axes = axes.flatten()

for i, feature in enumerate(features):
    axes[i].scatter(df_train[feature], df_train[y], alpha=0.7)
    axes[i].set_title(f'{feature} vs {y}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel(y)

for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
